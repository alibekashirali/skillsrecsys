## 1. Web-scrapping (Beatiful Soup)

```python
import requests
from bs4 import BeautifulSoup as bs
from time import sleep
from time import time
import re
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from numpy import random

import warnings
warnings.filterwarnings('ignore')

#скачивание html файлов со списком вакансии
def download_html(n):
    vac_num = []
    for i in range(1, n+1):
        sleep(1)
        res = requests.get(f'https://career.habr.com/vacancies?page={i}&sort=date&type=all')
        print(i, end=' ')
        with open(f'./parsing_new/htmls/habr_vac_{i}.html', 'w', encoding="utf-8") as f:
            f.write(res.text)
            

# скачивание html страниц вакансии
def download_pages(n):
    vac_num = []
    for k in range(1, n+1):
        with open(f'./parsing_new/htmls/habr_vac_{k}.html', 'r', encoding="utf-8") as f:
            text = f.read()
            
        print(k)
        url = bs(text, 'html.parser')
        url_a = url.find_all("a", {"class": "vacancy-card__title-link"})
        for p in url_a:
            a = p.get('href')[11:]
            if k % 2 == 0:
                sleep(random.randint(3))
            elif k % 3 == 0:
                sleep(random.rand())
            else:
                sleep(random.rand())
            vac_num.append(a)
            resp = requests.get('https://career.habr.com/vacancies/' + a)
            with open(f'./parsing_new/pages/habr_vac_{a}.html', 'w', encoding="utf-8") as l:
                l.write(resp.text)
    return vac_num

# парсинг данных с html страниц вакансии
def parse_html(page):
    info_array = []
    with open(f'./parsing_new/pages/habr_vac_{page}.html', 'r', encoding="utf-8") as f:
        text = f.read()
    soup = bs(text, 'html.parser')
    array = []
    info = {}
    info['title'] = soup.find('h1', 'page-title__title').text
    info['company'] = soup.find('div', 'company_name').text
    bag = soup.find_all('a', 'link-comp link-comp--appearance-dark')
    req = ""
    for i in bag:
        req = req + i.text + ", "
#         array.append(i.text)
    info['requirements'] = req
    info['description'] = soup.find('div', 'vacancy-description__text').text[14:]
    info['time'] = soup.find(attrs={"class": "basic-date"})['datetime']
    try:
        salary = soup.find('div', 'basic-salary basic-salary--appearance-vacancy-header').text.strip()

        if 'от' in salary:
            info['salary_min'] = re.search('от [ 0-9]+', salary).group().replace('от ', '').replace(' ', '')
        else:
            info['salary_min'] = None

        if 'до' in salary:
            info['salary_max'] = re.search('до [ 0-9]+', salary).group().replace('до ', '').replace(' ', '')
        else:
            info['salary_max'] = None

        if ('от' or 'до') in salary:
            info['salary_currency'] = re.search('[₽$€₸]', salary).group()
        else:
            info['salary_currency'] = None
    except:
        info['salary_min'], info['salary_max'], info['salary_currency'] = None, None, None

    info_array.append(info)
    return info_array
```

## 3. Preprocess Data

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")

def preprocess(text):
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words("english"))
    tokens = [token for token in tokens if token.isalnum() and token not in stop_words]
    stop_words = set(stopwords.words("russian"))
    tokens = [token for token in tokens if token.isalnum() and token not in stop_words]

    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    gen = ""
    for i in tokens:
      gen = gen + i + " "
    return gen
```

```python
df['skills'] = df['description']
df['skills'] = df['skills'].apply(preprocess)
```

```python
df['requirements'] = df['requirements'].apply(lambda x: x.split(", "))
```

```python
import PyPDF2

def pdftotext(name):
    with open(name, 'rb') as f:
        pdf_reader = PyPDF2.PdfReader(f)
        text = ''
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            page_text = page.extract_text()

            text += page_text
    return text
```

## 4. Skills Extraction

```python
import spacy
nlp = spacy.load("it_skills_ner_last_1")

def extract_skills(text):
    it_array = []
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == "IT-SKILL":
            it_array.append(ent.text)
    return it_array
```

```python
df['skills'] = df['skills'].apply(extract_skills)
```

### Clusterization using K-Means and TF-IDF for Job Title

```python
# Elbow Method

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

job_titles = df['title']
job_titles

job_titles_elbow = job_titles.to_list()

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(job_titles_elbow)

sse = []
max_k = 40
for k in range(1, max_k + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

plt.plot(range(1, max_k + 1), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

job_titles = df['title']

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(job_titles)

num_clusters = 30
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)


tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X.toarray())


colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'lime', 'orange', 'purple', 'gray', 'pink', 'olive', 
          'navy', 'maroon', 'brown', 'teal', 'indigo', 'violet', 'turquoise', 'gold', 'silver', 
          'darkgreen', 'coral', 'lavender', 'crimson', 'darkblue', 'lightblue', 'salmon', 'aquamarine', 
          'darkcyan', 'khaki', 'orchid', 'sienna', 'chartreuse', 'darkslategray', 'lightcoral', 'mediumorchid', 
          'palegreen', 'rosybrown', 'darkorange', 'darkviolet', 'hotpink', 'limegreen', 'mediumslateblue', 
          'paleturquoise', 'saddlebrown', 'darkkhaki']

plt.figure(figsize=(12, 8))

for i in range(num_clusters):
    cluster_points = X_tsne[kmeans.labels_ == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], c=colors[i], label=f'Cluster {i + 1}', alpha=0.7)

for i, job_title in enumerate(job_titles[0]):
    plt.annotate(job_title, (X_tsne[i, 0], X_tsne[i, 1]), fontsize=10, alpha=0.75)

plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.title('t-SNE Visualization of Clustered Job Titles')
plt.legend()
plt.show()
```

```python
def evaluate_recommendations(recommended_jobs, user_applications):
    # Количество выбранных вакансий, находящихся в рекомендациях
    selected_jobs = [job for job in recommended_jobs if job in user_applications]
    num_selected_jobs = len(selected_jobs)
    
    # Количество интересных пользователю вакансий
    num_interesting_jobs = len(user_applications)
    
    # Точность
    precision = num_selected_jobs / len(recommended_jobs)
    
    # Полнота
    recall = num_selected_jobs / num_interesting_jobs
    
    # F1-мера
    if precision + recall != 0:
        f1_score = 2 * (precision * recall) / (precision + recall)
    else:
        f1_score = 0.0

    # Охват
    coverage = len(set(recommended_jobs)) / num_interesting_jobs
    
    return precision, recall, f1_score, coverage

precision, recall, f1_score, coverage = evaluate_recommendations(rec_jobs, user_applications)

print("Precision:", precision)
print("Recall:", recall)
print("F1:", f1_score)
print("Coverage:", coverage)
```
